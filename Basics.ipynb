# Academy Award Nominations - A Dataquest guided project

### Column Headers

- `Year`  - the year of the awards ceremony.
- `Category` - the category of award the nominee was nominated for.
- `Nominee` - the person nominated for the award.
- `Additional Info` 
- `Won?` - this column contains either YES or NO.
- `Unnamed: 5`
- `Unnamed: 6` 
- `Unnamed: 7` 
- `Unnamed: 8` 
- `Unnamed: 9`
- `Unnamed: 10`

```python
import pandas as pd

data = pd.read_csv("academy_awards.csv", encoding="ISO-8859-1")

for i in range(5,11):
    print(data.iloc[:,i].value_counts())
    
print(data.columns)
    
data
```

```python
# Converting the Year column to just the year
data['Year'] = data['Year'].str[0:4]

# Change the Year column to ints
data['Year'] = data['Year'].astype("int64")

data
```

```python
data.dtypes
```

```python
#Selecting only the awards from after 2000
later_than_2000 = data[data['Year']>2000]
later_than_2000
```

```python
#Selecting only one of the four categories in the award_categories list
# and assigning them to nominations
award_categories = ['Actor -- Leading Role', 'Actor -- Supporting Role', 'Actress -- Leading Role', 'Actress -- Supporting Role']
nominations = later_than_2000[later_than_2000['Category'].isin(award_categories)]
nominations
```

```python
#Replacing the values in the won category from YES/NO to 1/0 to work with SQL
replace_dict = {'YES':1, 'NO':0}
nominations['Won?'] = nominations['Won?'].map(replace_dict)

final_nominations = nominations.drop(['Won?', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], axis=1)
final_nominations
```

```python
#Separating the Additional info column into 2 columns (Movie and Characters )
#Stripping right bracket and whitespace from the righthand side
additional_info_one = final_nominations['Additional Info'].str.rstrip("'}")

#Splitting the column on "{'" into a list that contains two values
additional_info_two = additional_info_one.str.split("{'")

final_nominations['Movies'] = additional_info_two.str[0]
final_nominations['Characters'] = additional_info_two.str[1]
final_nominations = final_nominations.drop(['Additional Info'], axis=1)

final_nominations

```

```python
#Writing to an SQL Database

import sqlite3

conn = sqlite3.connect("nominations.db")

final_nominations.to_sql("nominations", conn, index=False)
```

```python
query = "pragma table_info(nominations);"
query_2 = "select * from nominations limit 10"

schema = conn.execute(query).fetchall()
db = conn.execute(query_2).fetchall()

print(schema)
db
```
